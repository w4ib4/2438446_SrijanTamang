```python
def convert_length(value, direction):
    """
    Converts length between meters and feet.

    Parameters:
        value (float): The numeric value to convert.
        direction (str): 'm_to_ft' or 'ft_to_m'.

    Returns:
        float: The converted value.
    """
    if direction == 'm_to_ft':
        return value * 3.28084
    elif direction == 'ft_to_m':
        return value / 3.28084
    else:
        raise ValueError("Invalid direction. Use 'm_to_ft' or 'ft_to_m'.")


def convert_weight(value, direction):
    """
    Converts weight between kilograms and pounds.

    Parameters:
        value (float): The numeric value to convert.
        direction (str): 'kg_to_lbs' or 'lbs_to_kg'.

    Returns:
        float: The converted value.
    """
    if direction == 'kg_to_lbs':
        return value * 2.20462
    elif direction == 'lbs_to_kg':
        return value / 2.20462
    else:
        raise ValueError("Invalid direction. Use 'kg_to_lbs' or 'lbs_to_kg'.")


def convert_volume(value, direction):
    """
    Converts volume between liters and gallons.

    Parameters:
        value (float): The numeric value to convert.
        direction (str): 'l_to_gal' or 'gal_to_l'.

    Returns:
        float: The converted value.
    """
    if direction == 'l_to_gal':
        return value * 0.264172
    elif direction == 'gal_to_l':
        return value / 0.264172
    else:
        raise ValueError("Invalid direction. Use 'l_to_gal' or 'gal_to_l'.")


def unit_converter():
    """
    Main function to interactively convert between units of measurement.
    Supports length (m/ft), weight (kg/lbs), and volume (L/gal).
    Handles invalid inputs using try-except blocks.

    Returns:
        float or None: The converted result, or None if input is invalid.
    """
    print("Unit Converter")
    print("Choose conversion type:")
    print("1. Length (meters <-> feet)")
    print("2. Weight (kg <-> lbs)")
    print("3. Volume (liters <-> gallons)")

    choice = input("Enter 1, 2, or 3: ").strip()

    try:
        value = float(input("Enter the value to convert: "))
    except ValueError:
        print("Error: Please enter a valid numeric value.")
        return None

    try:
        if choice == '1':
            sub = input("Convert (1) m->ft or (2) ft->m? ")
            direction = 'm_to_ft' if sub == '1' else 'ft_to_m'
            result = convert_length(value, direction)
            unit_in = 'm' if sub == '1' else 'ft'
            unit_out = 'ft' if sub == '1' else 'm'
        elif choice == '2':
            sub = input("Convert (1) kg->lbs or (2) lbs->kg? ")
            direction = 'kg_to_lbs' if sub == '1' else 'lbs_to_kg'
            result = convert_weight(value, direction)
            unit_in = 'kg' if sub == '1' else 'lbs'
            unit_out = 'lbs' if sub == '1' else 'kg'
        elif choice == '3':
            sub = input("Convert (1) L->gal or (2) gal->L? ")
            direction = 'l_to_gal' if sub == '1' else 'gal_to_l'
            result = convert_volume(value, direction)
            unit_in = 'L' if sub == '1' else 'gal'
            unit_out = 'gal' if sub == '1' else 'L'
        else:
            print("Error: Unsupported conversion type.")
            return None

        print(f"Result: {value} {unit_in} = {result:.4f} {unit_out}")
        return result

    except ValueError as e:
        print(f"Conversion error: {e}")
        return None

print("Demo: convert_length(10, 'm_to_ft') =", convert_length(10, 'm_to_ft'))
print("Demo: convert_weight(70, 'kg_to_lbs') =", convert_weight(70, 'kg_to_lbs'))
print("Demo: convert_volume(5, 'l_to_gal') =", convert_volume(5, 'l_to_gal'))
```

```python
def compute_sum(numbers):
    """
    Calculates the sum of a list of numbers.

    Parameters:
        numbers (list of float): The input list.

    Returns:
        float: Sum of all elements.
    """
    if not numbers:
        raise ValueError("List is empty.")
    return sum(numbers)


def compute_average(numbers):
    """
    Calculates the average of a list of numbers.

    Parameters:
        numbers (list of float): The input list.

    Returns:
        float: Average of all elements.
    """
    if not numbers:
        raise ValueError("List is empty.")
    return sum(numbers) / len(numbers)


def compute_maximum(numbers):
    """
    Finds the maximum value in a list of numbers.

    Parameters:
        numbers (list of float): The input list.

    Returns:
        float: Maximum element.
    """
    if not numbers:
        raise ValueError("List is empty.")
    return max(numbers)


def compute_minimum(numbers):
    """
    Finds the minimum value in a list of numbers.

    Parameters:
        numbers (list of float): The input list.

    Returns:
        float: Minimum element.
    """
    if not numbers:
        raise ValueError("List is empty.")
    return min(numbers)


def math_operations():
    """
    Interactively performs sum, average, max, or min on a user-provided list.
    Handles invalid inputs and empty lists with try-except blocks.

    Returns:
        float or None: The computed result, or None on error.
    """
    print("Mathematical Operations")
    print("1. Sum  2. Average  3. Maximum  4. Minimum")
    choice = input("Choose operation (1-4): ").strip()

    try:
        raw = input("Enter numbers separated by spaces: ")
        numbers = [float(x) for x in raw.split()]
        if not numbers:
            raise ValueError("No numbers provided.")
    except ValueError as e:
        print(f"Input error: {e}")
        return None

    ops = {'1': compute_sum, '2': compute_average,
           '3': compute_maximum, '4': compute_minimum}
    names = {'1': 'Sum', '2': 'Average', '3': 'Maximum', '4': 'Minimum'}

    if choice not in ops:
        print("Invalid operation choice.")
        return None

    try:
        result = ops[choice](numbers)
        print(f"{names[choice]}: {result}")
        return result
    except ValueError as e:
        print(f"Error: {e}")
        return None

sample = [3.0, 7.5, 2.0, 9.0, 4.5]
print("Sample list:", sample)
print("Sum     :", compute_sum(sample))
print("Average :", compute_average(sample))
print("Maximum :", compute_maximum(sample))
print("Minimum :", compute_minimum(sample))
```

```python
# 1. Extract every other element
def extract_every_other(lst):
    """Returns every other element starting from index 0."""
    return lst[::2]

# 2. Slice a sublist (inclusive end)
def get_sublist(lst, start, end):
    """Returns elements from index start to end (inclusive)."""
    return lst[start:end + 1]

# 3. Reverse a list using slicing
def reverse_list(lst):
    """Returns a reversed copy of the list using slicing."""
    return lst[::-1]

# 4. Remove first and last elements
def remove_first_last(lst):
    """Returns the list without its first and last elements."""
    return lst[1:-1]

# 5. Get first n elements
def get_first_n(lst, n):
    """Returns the first n elements of the list."""
    return lst[:n]

# 6. Get last n elements
def get_last_n(lst, n):
    """Returns the last n elements of the list."""
    return lst[-n:]

# 7. Reverse skip: every second element starting from second-to-last, going backward
def reverse_skip(lst):
    """Returns every second element from second-to-last, moving backward."""
    return lst[-2::-2]
lst1 = [1, 2, 3, 4, 5, 6]
lst2 = [1, 2, 3, 4, 5]

print("extract_every_other([1,2,3,4,5,6]):", extract_every_other(lst1))   # [1,3,5]
print("get_sublist([1..6], 2, 4)         :", get_sublist(lst1, 2, 4))      # [3,4,5]
print("reverse_list([1,2,3,4,5])         :", reverse_list(lst2))           # [5,4,3,2,1]
print("remove_first_last([1,2,3,4,5])    :", remove_first_last(lst2))      # [2,3,4]
print("get_first_n([1,2,3,4,5], 3)       :", get_first_n(lst2, 3))         # [1,2,3]
print("get_last_n([1,2,3,4,5], 2)        :", get_last_n(lst2, 2))          # [4,5]
print("reverse_skip([1,2,3,4,5,6])       :", reverse_skip(lst1))           # [5,3,1]
```

```python
# 1. Flatten a nested list (one level deep)
def flatten(lst):
    """Flattens a one-level nested list into a single list."""
    return [item for sublist in lst for item in sublist]

# 2. Access nested element by a list of indices
def access_nested_element(lst, indices):
    """Returns the element in lst reached by following each index in indices."""
    element = lst
    for idx in indices:
        element = element[idx]
    return element

# 3. Sum of all elements in a nested list (any depth)
def sum_nested(lst):
    """Returns the sum of all numeric elements in a nested list of any depth."""
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

# 4. Remove specific element from a nested list
def remove_element(lst, elem):
    """Removes all occurrences of elem from each sub-list."""
    return [[x for x in sublist if x != elem] for sublist in lst]

# 5. Find maximum element in nested list (any depth)
def find_max(lst):
    """Returns the maximum element in a nested list of any depth."""
    flat = []
    def _flatten(l):
        for item in l:
            if isinstance(item, list):
                _flatten(item)
            else:
                flat.append(item)
    _flatten(lst)
    return max(flat)

# 6. Count occurrences of an element in a nested list
def count_occurrences(lst, elem):
    """Counts how many times elem appears in a nested list."""
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count

# 7. Deep flatten (any depth)
def deep_flatten(lst):
    """Flattens a nested list of any depth into a single flat list."""
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

# 8. Average of all elements in a nested list
def average_nested(lst):
    """Calculates the average of all elements in a nested list."""
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)

print("flatten([[1,2],[3,4],[5]])            :", flatten([[1,2],[3,4],[5]]))
print("access_nested_element, indices [1,2] :", access_nested_element([[1,2,3],[4,5,6],[7,8,9]], [1,2]))
print("sum_nested([[1,2],[3,[4,5]],6])       :", sum_nested([[1,2],[3,[4,5]],6]))
print("remove_element([[1,2],[3,2],[4,5]],2):", remove_element([[1,2],[3,2],[4,5]], 2))
print("find_max([[1,2],[3,[4,5]],6])         :", find_max([[1,2],[3,[4,5]],6]))
print("count_occurrences([[1,2],[2,3],[2,4]],2):", count_occurrences([[1,2],[2,3],[2,4]], 2))
print("deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]):", deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]))
print("average_nested([[1,2],[3,4],[5,6]])   :", average_nested([[1,2],[3,4],[5,6]]))
```

```python
import numpy as np

A = np.array([[3, 4], [7, 8]], dtype=float)
B = np.array([[5, 3], [2, 1]], dtype=float)

A_inv = np.linalg.inv(A)

# 1. Prove A * A^-1 = I
print("1. A @ A_inv (should be Identity):\n",
      np.round(A @ A_inv, decimals=10))

# 2. Prove AB != BA
AB = A @ B
BA = B @ A
print("\n2. AB =\n", AB)
print("   BA =\n", BA)
print("   AB == BA:", np.allclose(AB, BA))  # Should be False

# 3. Prove (AB)^T = B^T * A^T
lhs = (A @ B).T
rhs = B.T @ A.T
print("\n3. (AB)^T =\n", lhs)
print("   B^T @ A^T =\n", rhs)
print("   Equal:", np.allclose(lhs, rhs))  # Should be True

# ---- Solve system of linear equations ----
# 2x - 3y + z  = -1
# x  - y  + 2z = -3
# 3x + y  - z  =  9
coef = np.array([[ 2, -3,  1],
                 [ 1, -1,  2],
                 [ 3,  1, -1]], dtype=float)
b_vec = np.array([-1, -3, 9], dtype=float)

# Inverse method: X = A^-1 * B
coef_inv = np.linalg.inv(coef)
X_inv = coef_inv @ b_vec
print("\n--- System of Equations ---")
print("Solution via Inverse method (x, y, z):", np.round(X_inv, 4))

# Using np.linalg.solve
X_solve = np.linalg.solve(coef, b_vec)
print("Solution via linalg.solve  (x, y, z):", np.round(X_solve, 4))

# Verify
print("Verification A @ X =", np.round(coef @ X_solve, 4), " (should be", b_vec, ")")
```
